**System of Equations**

```python
# coefficient matrix
a=np.random.randint(1,11,[5,5])
#dependent variable
b=np.array([2,5,8,9,7])
x=np.linalg.solve(a,b)
print('Coefficient Matrix')
print(a)
print('\nDependent Variable')
print(b)
print('\nSolution')
print(x)

# check correctness
A=a.dot(x)
print(A)
```

**Singular Value Decomposition**

```python
from skimage import data
image=data.coins()
print(image.dtype,image.min(),image.max(),image.shape)
plt.imshow(image)
plt.show()
```

```python
img=image/252.0
img=np.transpose(img,axes=(1,0))
```

```python
A,B,C=np.linalg.svd(img)
print(A.shape,B.shape,C.shape)
```

```python
Sigma=np.zeros((303,303,3))
for i in range(3):
  np.fill_diagonal(Sigma[:,:,i],B[i])
print(Sigma.shape)
```

```python
# reconstructing original image
Sigma=np.zeros((384,303))
np.fill_diagonal(Sigma,B)
reconst = A @ Sigma @ C
reconst=np.transpose(reconst,axes=(1,0))
plt.imshow(reconst)
plt.show()
```

```python
#SVD reconstruction
k=30
reconst=A @ Sigma [:,:k] @ C[:k,:]
reconst=np.transpose(reconst,axes=(1,0))
plt.imshow(reconst)
plt.show()
```
